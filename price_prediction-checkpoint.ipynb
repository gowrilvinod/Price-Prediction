```python
data['pop_count'] = np.ceil(data['median_income']/1.5)
data['pop_count'].where(data['pop_count']<5, 5.0, inplace=True)
data['pop_count'].hist()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(data,data['pop_count']):
    strat_train_index = data.loc[train_index]
    strat_test_index = data.loc[test_index]

strat_test_index.head(8)
```
